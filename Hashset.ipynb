```python
import matplotlib.pyplot as plt
import numpy as np

x_random = np.array([5, 10,20,30,50,80,100,150,200,500])
y_random = np.array([0.253, 0.253, 0.281, 0.331, 0.361, 0.409, 0.485, 0.514, 0.678, 1.244])

x_sort = np.array([5, 10,20,30,50,80,100,150,200,500])
y_sort = np.array([0.255, 0.262, 0.287, 0.316, 0.364, 0.406, 0.470, 0.548, 0.677, 1.284])

x_reverse = np.array([5, 10,20,30,50,80,100,150,200,500])
y_reverse = np.array([0.253, 0.266, 0.286, 0.300, 0.376, 0.406, 0.485, 0.552, 0.672, 1.271])


plt.plot(x_random, y_random, label= "random" )
plt.xlabel("Size of Dictionary (in thousands)")
plt.ylabel("Time (in seconds)")
plt.legend()

plt.show()
```

```python
plt.plot(x_sort, y_sort, label  = "sorted")
plt.xlabel("Size of Dictionary (in thousands)")
plt.ylabel("Time (in seconds)")
plt.legend()

```

```python

plt.plot(x_reverse, y_reverse, label = "reverse sorted")
plt.xlabel("Size of Dictionary (in thousands)")
plt.ylabel("Time (in seconds)")

plt.legend()
```
